## 3.3 - USE (Universal Sentence Encoder)
- USE capture le sens global d’une phrase, contrairement aux word embeddings classiques.
- Impact : USE améliore la compréhension du contexte global d’un tweet.
### 3.3.1 - Vectorisation

```python
import tensorflow_hub as hub
```

```python
vectorized_path = "tweets_vectorized.parquet"
```

```python
if os.path.exists(vectorized_path):
    print("✅ Chargement des embeddings USE déjà vectorisés...")
    tweets = pd.read_parquet(vectorized_path)
    print("✅ Chargement terminé.")
else:
    print("🔄 Aucun fichier vectorisé trouvé. Vectorisation en cours...")
    
    # Chargement du modèle USE
    use_model = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")
    
    # Transformation des tweets
    tweets['use_embedding'] = tweets['text'].apply(lambda x: use_model([x]).numpy().flatten())
    
    # Sauvegarde
    tweets.to_parquet(vectorized_path, index = False)
    print("✅ Vectorisation terminée et sauvegardée sous tweets_vectorized.parquet !")
```

### 3.3.2 - Evaluation USE

```python
X_use_full = np.vstack(tweets['use_embedding'].values)
X_use_train, X_use_test, y_train, y_test = train_test_split(X_use_full, tweets['label'], test_size = 0.2, random_state = 70, stratify = tweets['label'])
```

```python
log_reg_use = LogisticRegression(max_iter=200)
log_reg_use.fit(X_use_train, y_train)
y_pred_use = log_reg_use.predict(X_use_test)
print(classification_report(y_test, y_pred_use))
functions.plot_confusion_matrix(y_test, y_pred_use, "Régression Logistique (USE)")
```

### 3.3.3 - LightGBM (USE)

### 5.2.2 - FastText + LSTM

```python
lstm_eval_results = joblib.load("lstm_eval_simulation.pkl")
results['Modèle'].append('FastText + LSTM')
results['Accuracy'].append(round(lstm_eval_results['accuracies'][-1], 4))
results['F1-score'].append("≈ à recalculer manuellement")  # (Pas sauvegardé pour l'instant)
results['Temps entraînement'].append("Long (~4h)")
results['Ressources'].append("CPU")
```

### 5.2.3 - USE + LightGBM

```python
# Chargement des embeddings USE et split identique
X_use_full = np.vstack(tweets['use_embedding'].values)
y_labels = tweets['label'].astype(int).values

X_use_train, X_use_test, y_train_use, y_test_use = train_test_split(
    X_use_full, y_labels, test_size=0.2, random_state=70, stratify=y_labels
)

# Chargement du modèle LightGBM sauvegardé
lgbm_model = lgb.Booster(model_file="lgbm_model.txt")

# Prédictions
y_pred_lgbm = (lgbm_model.predict(X_use_test) > 0.5).astype(int)

# Ajout des résultats
results['Modèle'].append('USE + LightGBM')
results['Accuracy'].append(round(accuracy_score(y_test_use, y_pred_lgbm), 4))
results['F1-score'].append(round(f1_score(y_test_use, y_pred_lgbm), 4))
results['Temps entraînement'].append("2 min")
results['Ressources'].append("CPU")
```

### 5.2.4 - FastText + RandomForest

```python
# Chargement des embeddings FastText + labels
X_fasttext_full = np.vstack(tweets['fasttext_embedding'].values)
y_labels = tweets['label'].astype(int).values

# Split identique à l'entraînement
X_ft_train, X_ft_test, y_train_ft, y_test_ft = train_test_split(
    X_fasttext_full, y_labels, test_size = 0.2, random_state = 70, stratify = y_labels
)

# Chargement du modèle RandomForest déjà sauvegardé
rf_model = joblib.load("rf_model.pkl")

# Prédiction
y_pred_rf = rf_model.predict(X_ft_test)

# 🔥 Correction : Conversion en int
y_pred_rf = y_pred_rf.astype(int)

# Ajout des résultats
results['Modèle'].append('FastText + RandomForest')
results['Accuracy'].append(round(accuracy_score(y_test_ft, y_pred_rf), 4))
results['F1-score'].append(round(f1_score(y_test_ft, y_pred_rf), 4))
results['Temps entraînement'].append("5 min")
results['Ressources'].append("CPU")
```
