## 3.3 - USE (Universal Sentence Encoder)
- USE capture le sens global dâ€™une phrase, contrairement aux word embeddings classiques.
- Impact : USE amÃ©liore la comprÃ©hension du contexte global dâ€™un tweet.
### 3.3.1 - Vectorisation

```python
import tensorflow_hub as hub
```

```python
vectorized_path = "tweets_vectorized.parquet"
```

```python
if os.path.exists(vectorized_path):
    print("âœ… Chargement des embeddings USE dÃ©jÃ  vectorisÃ©s...")
    tweets = pd.read_parquet(vectorized_path)
    print("âœ… Chargement terminÃ©.")
else:
    print("ğŸ”„ Aucun fichier vectorisÃ© trouvÃ©. Vectorisation en cours...")
    
    # Chargement du modÃ¨le USE
    use_model = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")
    
    # Transformation des tweets
    tweets['use_embedding'] = tweets['text'].apply(lambda x: use_model([x]).numpy().flatten())
    
    # Sauvegarde
    tweets.to_parquet(vectorized_path, index = False)
    print("âœ… Vectorisation terminÃ©e et sauvegardÃ©e sous tweets_vectorized.parquet !")
```

```python
X_use_full = np.vstack(tweets['use_embedding'].values)
X_use_train, X_use_test, y_train, y_test = train_test_split(X_use_full, tweets['label'], test_size = 0.2, random_state = 70, stratify = tweets['label'])
```

```python
log_reg_use = LogisticRegression(max_iter=200)
log_reg_use.fit(X_use_train, y_train)
y_pred_use = log_reg_use.predict(X_use_test)
print(classification_report(y_test, y_pred_use))
functions.plot_confusion_matrix(y_test, y_pred_use, "RÃ©gression Logistique (USE)")
```

```python
fig, ax1 = plt.subplots(figsize=(12,6))

color = 'tab:blue'
ax1.set_xlabel('Training Steps')
ax1.set_ylabel('Train Loss', color=color)
ax1.plot(range(1, len(train_loss)+1), train_loss, color=color, label='Train Loss')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # DeuxiÃ¨me axe
color = 'tab:orange'
ax2.set_ylabel('Eval Loss', color=color)
ax2.plot(eval_steps, eval_loss, marker='o', color=color, label='Eval Loss')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.title("DistilBERT - Loss Evolution (Train & Eval)")
plt.show()
```
